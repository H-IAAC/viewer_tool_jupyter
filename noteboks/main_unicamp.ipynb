```python
def read_csv_exp(csv_file_name):
    df = pd.read_csv(csv_file_name)
    df['sec'] = [x / 1000 for x in df['VideoTimelapse']]
    return df    
```

```python
## todos os arquivos
exp='TS001'
activity='all'
user="01"
```

```python
plot_1_name='LinearAccelerometer_Anotador'
csv_file_name = f'Data/{exp}/{user}/{activity}/{plot_1_name}.csv'
data_plot_1=read_csv_exp(csv_file_name)
data_plot_1.head()
```

```python
plot_2_name='Accelerometer_Bolso Direito'
csv_file_name = f'Data/{exp}/{user}/{activity}/{plot_2_name}.csv'
data_plot_2=read_csv_exp(csv_file_name)
data_plot_2.head()
```

```python
app = DashboardApp()

player = VideoPlayer(
    app,
    video_path=f"../noteboks/Data/{exp}/{user}/{activity}/video_1948708557748126639.mp4",
    volume=0.0,
)
controller = Controller(app)
controller.connectToVideo(player)

plot_1 = LinearPlot(
    app,
    data_df=data_plot_1,
    x_col="sec",
    y_cols=["Value 1", "Value 2", "Value 3"],
    max_points=500,
    id="plot_1",
    title=plot_1_name,
)

plot_2 = LinearPlot(
    app,
    data_df=data_plot_2,
    x_col="sec",
    y_cols=["Value 1", "Value 2", "Value 3"],
    max_points=500,
    id="plot_2",
    title=plot_2_name,
)

controller.addGraph(plot_1)
controller.addGraph(plot_2)

horizontal_layout = HorizontalLayout()
horizontal_layout.append(player)
horizontal_layout.append(controller)
app.add_component(horizontal_layout)

app.run(port=8050)
```
